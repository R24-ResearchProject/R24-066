```python
from ultralytics import YOLO
import cv2
import os
```

```python
model = YOLO("/content/runs/segment/train/weights/best.pt")
```

```python
video_path = "/content/Test1.MOV"
```

```python
cap = cv2.VideoCapture(video_path)
```

```python
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
```

```python
# Define the codec and create VideoWriter object to save output video
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
output_path = "/content/annotated_output.mp4"
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))
```

```python
# Process each frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Run inference on the frame
    results = model(frame)

    # Draw bounding boxes and labels on the frame
    annotated_frame = results[0].plot()

    # Write the annotated frame to the output video
    out.write(annotated_frame)
```

```python
# Release the video capture and writer objects
cap.release()
out.release()

# Step 6: Display the Results
# Function to display the video within Google Colab
from IPython.display import HTML
from base64 import b64encode
```

```python
def display_video(video_path):
    mp4 = open(video_path,'rb').read()
    data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
    return HTML(f'<video width="640" height="480" controls><source src="{data_url}" type="video/mp4"></video>')

# Display the output video
display_video(output_path)
```
