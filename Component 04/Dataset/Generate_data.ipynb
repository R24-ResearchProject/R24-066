```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
from faker import Faker

# Define the number of employees and total records in the dataset
num_employees = 1000
total_records = 5000

# Create a Faker instance
fake = Faker()

# Generate names without "Mr." or "Miss"
def generate_name():
    name = fake.name()
    while name.startswith("Mr.") or name.startswith("Miss"):
        name = fake.name()
    return name

# Generate names
names = [generate_name() for _ in range(num_employees)]

# Generate ID numbers
id_numbers = [random.randint(10000, 99999) for _ in range(num_employees)]

# Generate joining dates
start_date = datetime(2020, 1, 1)
end_date = datetime(2024, 3, 31)
joining_dates = [start_date + timedelta(days=random.randint(0, (end_date - start_date).days)) for _ in range(num_employees)]

# Generate ages
min_age = 20
max_age = 30
ages = [random.randint(min_age, max_age) for _ in range(num_employees)]

# Generate birth dates from ages
present_date = datetime.now()
birth_dates = [present_date - timedelta(days=365*age) for age in ages]

# Generate genders
genders = ['Female']
gender_probabilities = [1.0]  # Only female employees
genders = np.random.choice(genders, size=num_employees, p=gender_probabilities)

# Generate average working hours
min_hours = 20
max_hours = 50
average_working_hours = [random.uniform(min_hours, max_hours) for _ in range(num_employees)]

# Generate Education level
education = [random.choice(['O/L', 'A/L', 'Diploma']) for _ in range(num_employees)]

# Determine skill levels based on joining dates and average working hours
skill_levels = []
for joining_date, avg_hours in zip(joining_dates, average_working_hours):
    days_since_joining = (present_date - joining_date).days
    if days_since_joining > 365 * 5 or avg_hours > 40:  # More than 5 years or avg_hours > 40
        skill_levels.append(random.choice(['Expert']))
    else:
        skill_levels.append(random.choice(['Beginner', 'Intermediate']))

# Generate defect types and defect rates
defect_types = ['Runoff', 'Openseam', 'SPI Errors', 'High Low']
defect_rates = [random.uniform(0, 1) for _ in range(total_records)]
defect_data = [{'Type': defect_type, 'Rate': defect_rate} for defect_type, defect_rate in zip(np.random.choice(defect_types, size=total_records), defect_rates)]

# Select a random set of 4 defect types for each employee
employee_defects = [random.sample(defect_types, 4) for _ in range(num_employees)]

# Create DataFrame
data = {
    'ID': [id for _ in range(total_records) for id in [id_numbers[_ // (total_records // num_employees)]]],
    'Name': [name for _ in range(total_records) for name in [names[_ // (total_records // num_employees)]]],
    'Joining Date': [date for _ in range(total_records) for date in [joining_dates[_ // (total_records // num_employees)]]],
    'Age': [age for _ in range(total_records) for age in [ages[_ // (total_records // num_employees)]]],
    'Birth Date': [bd for _ in range(total_records) for bd in [birth_dates[_ // (total_records // num_employees)]]],
    'Gender': [gender for _ in range(total_records) for gender in [genders[_ // (total_records // num_employees)]]],
    'Education': [edu for _ in range(total_records) for edu in [education[_ // (total_records // num_employees)]]],
    'Average Working Hours': [hours for _ in range(total_records) for hours in [average_working_hours[_ // (total_records // num_employees)]]],
    'Skill Level': [skill for _ in range(total_records) for skill in [skill_levels[_ // (total_records // num_employees)]]],
    'Defect Type': [defect_set[random.randint(0, 3)] for defect_set in employee_defects for _ in range(total_records // num_employees)],
    'Defect Rate': ["{:.4f}".format(defect_data[_]['Rate']) for _ in range(total_records)],
}

# Add weekday column
data['Weekday'] = [(date.weekday() + 1) % 7 + 1 for date in data['Joining Date']]

df = pd.DataFrame(data)

# Display the first few rows of the DataFrame
print(df.head())

# Save data CSV
# df.to_csv('demographic_data_update.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Define the defect types
defect_types = ['Runoff', 'Openseam', 'SPI Errors', 'High Low']

# Generate 100 employee IDs
employee_ids = ['EMP' + str(i).zfill(3) for i in range(1, 101)]

# Generate dates from 2024-01-01 to today
start_date = datetime(2024, 1, 1)
end_date = datetime.now()
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# Initialize an empty list to store the data
data = []

# Generate data for each employee
for _ in range(100):
    emp_id = random.choice(employee_ids)
    date = random.choice(date_range)
    defects = random.sample(defect_types, random.randint(1, len(defect_types)))
    for defect in defects:
        defect_count = random.randint(50, 600)
        defect_rate = random.uniform(0.0, 100.0)  # Random defect rate between 0.0 and 100.0
        weekday = date.strftime('%A')
        data.append({'Employee_ID': emp_id, 'Date': date, 'Defect_Type': defect, 
                     'Defect_Count': defect_count, 'Defect_Rate': defect_rate, 'Weekday': weekday})

# Create a DataFrame from the generated data
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Define the defect types
defect_types = ['Runoff', 'Openseam', 'SPI Errors', 'High Low']

# Generate 250 employee IDs
employee_ids = ['EMP' + str(i).zfill(3) for i in range(1, 251)]

# Generate dates from 2024-01-01 to today
start_date = datetime(2024, 1, 1)
end_date = datetime.now()
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# Initialize an empty list to store the data
data = []

# Generate data for each employee ID and date combination
for emp_id in employee_ids:
    for date in date_range:
        # Shuffle the defect types for each date
        random_defect_types = random.sample(defect_types, len(defect_types))
        # Calculate defect count and defect rate for each defect type
        for i, defect in enumerate(random_defect_types, start=1):
            defect_count = random.randint(50, 600)
            defect_rate = random.uniform(0.0, 100.0)  # Random defect rate between 0.0 and 100.0
            weekday = date.strftime('%A')
            data.append({'Employee_ID': emp_id, 'Date': date, 'Defect_Type': defect, 
                         'Defect_Count': defect_count, 'Defect_Rate': defect_rate, 'Weekday': weekday})
            # Break the loop if all 4 defect types have been added
            if i == 4:
                break

# Create a DataFrame from the generated data
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
print(df.tail())
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Define the defect types
defect_types = ['Runoff', 'Openseam', 'SPI Errors', 'High Low']

# Generate 10 unique employee IDs
employee_ids = random.sample(['EMP' + str(i).zfill(3) for i in range(1, 1001)], 10)

# Generate dates from 2024-01-01 to today
start_date = datetime(2024, 1, 1)
end_date = datetime.now()
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

# Initialize an empty list to store the data
data = []

# Generate data for each employee ID and date combination
for emp_id in employee_ids:
    for date in date_range:
        # Skip if the date is not a weekday (Monday to Friday)
        if date.weekday() < 5:
            # Shuffle the defect types for each date
            random_defect_types = random.sample(defect_types, len(defect_types))
            # Calculate defect count and defect rate for each defect type
            for defect in random_defect_types:
                defect_count = random.randint(50, 100)
                defect_rate = random.uniform(0.0, 100.0)  # Random defect rate between 0.0 and 100.0
                weekday = date.strftime('%A')
                data.append({'Employee_ID': emp_id, 'Date': date, 'Defect_Type': defect, 
                             'Defect_Count': defect_count, 'Defect_Rate': defect_rate, 'Weekday': weekday})

# Create a DataFrame from the generated data
df = pd.DataFrame(data[:250])  # Take the first 250 records

# Display the DataFrame
print(df)

# Save data CSV
df.to_csv('time_series_data_01.csv', index=False)
```
